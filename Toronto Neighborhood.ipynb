```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
data_frame = pd.DataFrame()
for i in range(9):
    data_frame[i] = df[i].apply(lambda x: x)
```

```python
column_names = ['Postcode','Borough', 'Neighbourhood']
data_frame_new = pd.DataFrame(columns = column_names)
```

```python
lst_borough = []
lst_neighbor = []
pst_cd = []

def func(x):
    global lst_borough, lst_neighbor, pst_cd
    post_code = x[:3]
    Borough = x[3:].split('(')[0]
    Neighbor = x[3:].split('(')

    if len(Neighbor) > 1:
        neighbor_lst = Neighbor[1].split(')')[0]
        neighbor_lst = neighbor_lst.split('/')
        for i in range(len(neighbor_lst)):
            
            lst_borough.append(Borough)
            neighbor = neighbor_lst[i]
            lst_neighbor.append(neighbor)
            pst_cd.append(post_code)
           
        return True
    else :
       
        lst_borough.append(Borough)
        lst_neighbor.append('Not assigned')
        pst_cd.append(post_code)
        
        return False
    
ser1.apply(func)
```

```python
data_frame_new
```

```python
df = pd.concat([pd.DataFrame(pst_cd),pd.DataFrame(lst_borough), pd.DataFrame(lst_neighbor)], axis = 1)
df.columns = ['Postcode','Borough', 'Neighbourhood'] 
```

```python
df
```

### Data processing and cleaning

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
df2.shape
```

### Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

### Merging dataframe on post code to get lattitude and longitude for neighbors in Canada

```python
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postcode')
df3.head()
```

### Clustering and plotting neighborhoods in Toronto Canada

### Filter out rows without Toronto in Borough

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

### Using Folium we are visualizing all the neighborhood in Toronto

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Clustering neighborhhods in Toronto using k-means

```python
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
